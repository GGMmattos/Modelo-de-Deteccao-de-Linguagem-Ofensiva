```python
base_hatebr = pd.read_csv('/content/drive/MyDrive/PIC /Dataset/HateBR.csv')
told_br = pd.read_csv("/content/drive/MyDrive/PIC /Dataset/told_binary_2_colums.csv")

base_hatebr_train_data  = base_hatebr.iloc[:,[0, 1]]
told_br_eval_data = told_br.iloc[:,[0, 1]]


train_df = pd.DataFrame({
    'text': base_hatebr_train_data['instagram_comments'],
    'label': base_hatebr_train_data['offensive_language']
})

eval_df = pd.DataFrame({
    'text': told_br_eval_data['text'],
    'label': told_br_eval_data['toxic']
})

model_args = ClassificationArgs(num_train_epochs=1)

model = ClassificationModel(
    "bert", "neuralmind/bert-base-portuguese-cased", args=model_args,
    use_cuda=False
)

model.train_model(train_df)

with open('Hate_BR-BERT', 'wb') as file:
    pickle.dump(model, file)
```

```python
with open('/content/drive/MyDrive/PIC /Models/Hate_BR-BERT_FULL.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
lst = []
for arr in model_outputs:
    lst.append(np.argmax(arr))
true = eval_df['label'].tolist()
predicted = lst
print(classification_report(true, predicted))
```
